Print i as long as i is less than 7:

```python
i = 1
while i < 7:
  print(i)
  i += 1
```

Exit the loop when i is 4:

```python
i = 1
while i < 7:
  print(i)
  if i == 4:
    break
  i += 1
```

Print each fruit in a color list:

```python
fruits = ["red", "black", "green"]
for x in fruits:
  print(x)
```

Loop through the letters in the word "datainsight":

```python
for x in "datainsight":
  print(x)
```
